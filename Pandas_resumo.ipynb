#Uso de planilhas em excel

````python
df1 = pd.read_excel(path_root + "Aracaju.xlsx")
df2 = pd.read_excel(path_root +"Fortaleza.xlsx")
df3 = pd.read_excel(path_root +"Natal.xlsx")
df4 = pd.read_excel(path_root +"Recife.xlsx")
df5 = pd.read_excel(path_root +"Salvador.xlsx")
````

````python
df1.head()
````

````python
#juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
df.head()
````

````python
df.sample(5)
````

````python
df.dtypes
````

````python
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
````

###tratando valores faltantes:
- Substituindo os valores nulos por zero 

```
df["Vendas"].fillna(0, inplace=True)
```
- Apagando as linhas com valores nulos

```
df.dropna(inplace=True)
```
- Apagando as linhas com valores nulos com base apenas em 1 coluna

```
df.dropna(subset=["Vendas"], inplace=True)
```
- Removendo linhas que estejam com valores faltantes em todas as colunas

```
df.dropna(how="all", inplace=True)
```









````python
#Consultando linhas com valores faltantes
df.isnull().sum()
````

###Criando colunas

````python
#Criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
````

````python
df["Receita/Vendas"] = df["Receita"] / df["Vendas"] 
df.head()
````

#Trabalhando com Datas

````python
#Trasnformando a coluna de data em tipo inteiro
df["Data"] = df["Data"].astype("int64")
````

````python
#Verificando o tipo de dado de cada coluna
df.dtypes
````

````python
#Transformando coluna de data em data
df["Data"] = pd.to_datetime(df["Data"])
df.dtypes
````

````python
#Agrupamento por ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
````

````python
#Criando uma nova coluna com o ano
df["Ano_Venda"] = df["Data"].dt.year
df.sample(5)
````

````python
#Extraindo o mês e o dia
df["mes_venda"], df["dia_venda"] = (df["Data"].dt.month, df["Data"].dt.day)
````

````python
#Retornando a data mais antiga
df["Data"].min()
````

````python
#Calculando a diferença de dias
df["diferenca_dias"] = df["Data"] - df["Data"].min()
df.sample(5)
````

````python
#Criando a coluna de trimestre
df["trimestre_venda"] = df["Data"].dt.quarter
df.sample(5)
````

````python
#Filtrando as vendas de 2019 do mês de março
vendas_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
vendas_marco_19.sample(20)
````

#Análise exploratória

````python
df = pd.read_excel(path_root + "AdventureWorks.xlsx")
df.head()
````

````python
df.shape
````

````python
df.dtypes
````

````python
#Qual a Receita total?
df["Valor Venda"].sum()
````

````python
#Qual o custo Total?
df["custo"] = df["Custo Unitário"]*(df["Quantidade"]) #Criando a coluna de custo
df.head(1)
````

````python
#Qual o custo Total?
round(df["custo"].sum(), 2)
````

````python
#Agora que temos a receita e custo e o total, podemos achar o Lucro total
#Vamos criar uma coluna de Lucro que será Receita - Custo
df["lucro"]  = df["Valor Venda"] - df["custo"]
df.head()
````

````python
#Total Lucro
round(df["lucro"].sum(),2)
````

````python
#Criando uma coluna com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
df.head(1)
````

````python
#Extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
df.head(1)
````

````python
#Verificando o tipo da coluna Tempo_envio
df["Tempo_envio"].dtype
````

````python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo_envio"].mean()
````

````python
#Verificando se temos dados faltantes
df.isnull().sum()
````
